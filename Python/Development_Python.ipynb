## BMI CALCULATOR

https://www.calculator.net/bmi-calculator.html?cage=25&csex=m&cheightfeet=5&cheightinch=10&cpound=160&cheightmeter=180&ckg=65&printit=0&ctype=standard

```python
# BMI = (Weight in pounds x 703) / (height in inches x height in inches)
```

```python
name = input('Enter your name: ')
weight = float(input('Enter your weight in pounds: '))
height = float(input('Enter your height in inches: '))

BMI = (weight * 703)/ (height * height)
print(BMI)

if BMI>0:
    if(BMI < 18.5):
        print(name + ', You are underweight.')
    elif(BMI <= 24.9):
        print(name + ', You are normal weight.')
    elif(BMI <= 29.9):
        print(name + ', You are overweight.')
    elif(BMI <= 34.9):
        print(name + ', You are obese.')
    elif(BMI <= 39.9):
        print(name + ', You are severely obese.')
    elif(BMI > 40):
        print(name + ', You are Mordibly obese.')
    else:
        print('Enter valid inputs')
```

## AUTOMATIC FILE SORTER IN FILE EXPLORER

```python
import os, shutil
```

```python
path = r"C:/Users/santi/OneDrive/Desktop/Fotos LMA/"
```

```python
file_name = os.listdir(path)
print(file_name)
```

```python
folder_names = ['fotos', 'videos']

for loop in range(0, 2):
    if not os.path.exists(path + folder_names[loop]):
        os.makedirs(path + folder_names[loop])
```

```python
for file in file_name:
    if (".mp4" in file or ".MOV" in file) and not os.path.exists(path + "videos/" + file):
        shutil.move(path + file, path + "videos/" + file)
    elif (".jpg" in file or ".jpeg" in file or ".JPG" in file) and not os.path.exists(path + "fotos/" + file):
        shutil.move(path + file, path + "fotos/" + file)
    else:
        print("There are files in this path that were not moved!")
```
